### i. Libraries and Dependencies

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
% matplotlib inline
```

Example: Historic Weather Monthly Averages in San Francisco

```python
average_lows = [46, 48, 49, 49, 51, 53, 54, 55, 55, 54, 50, 54]
average_highs = [57, 60, 62, 63, 64, 67, 67, 68, 70, 69, 63, 57]
```

```python
months = [date.strftime("%b") \
          for date in pd.date_range(start='1/2017', end='1/2018', freq='M')]
months
```

#### concatenate to the dataframe from list
![Creating_data_frame_from_List/Dictionary](http://pbpython.com/images/pandas-dataframe-shadow.png)

```python
data={ 'month': months, 'average_lows': average_lows, 'average_highs':average_highs }
```

```python
df0=pd.DataFrame(data)
df0.head()
```

```python
#OR simply use pd.DataFrame
df= pd.DataFrame({'month': months,
                 'average_low':average_lows,
                 'average_high':average_highs})
df.head()
```

```python
df=df.set_index('month')
```

### ii. Plot Monthly Weather Data

**1. Version 0**

```python
df.plot()
```

**Version 1**  

**Requirements:**
><li> Add grid | Add dots 
><li> Figsize (8,7)  
><li> Add title on the graph (Title fontsize=18 |  pad=1.01 )  
><li> X lable ("Month" | fontsize=13 | lable pad= 15)  
><li> X tiks ('Each month')
><li> Y lable ("Temperature (degrees Fahrenheit)" | fontsize=14 | lable pad= 15)  
><li> Legend  (fontsize=13)
    
  

```python
df.index
```

```python
df.plot(kind='line', grid=True, figsize=(10,7), style='o-')
plt.title('Historical Yearly Averages in San Francisco', fontsize=18, y=1.01)
plt.xlabel('Month', fontsize=13, labelpad=15)
plt.xticks(range(12),df.index)
plt.ylabel('Temperature (degrees Fahrenheit)', fontsize=14, labelpad=15)
plt.legend(fontsize=13);
```

**Version 2**  

**Requirements:**
> <li> include `average_yearly_high` and `average_yearly_low` too - each as a dotted line  
> <li> x-axis ticks each represent the months of the years, starting in January and ending in December 
> <li> y-axis should be Temperature (degrees Fahrenheit)

```python
df['yearly_average_high']=sum(df.average_high)/12
```

```python
df['yearly_average_low']=sum(df.average_low)/12
```

```python
df.head()
```

```python
df.yearly_average_high.plot(kind='line', grid=True, figsize=(15,10), style='--',color='blue')
df.yearly_average_low.plot(kind='line', grid=True, figsize=(15,10), style='--',color='green')
df.average_high.plot(kind='line', grid=True, figsize=(15,10), linewidth=1.5,style='o-')
df.average_low.plot(kind='line', grid=True, figsize=(15,10),linewidth=1.5, style='o-')
sns.set_context('poster')
plt.title('Historical Yearly Averages in San Francisco', fontsize=18, y=1.01)
plt.xlabel('Month', fontsize=15, labelpad=15)
plt.xticks(range(12),df.index,fontsize=16, rotation=25)
plt.yticks(fontsize=16)
plt.ylabel('Temperature (degrees Fahrenheit)', fontsize=14, labelpad=15)
plt.legend(fontsize=13,loc=2); 
```
